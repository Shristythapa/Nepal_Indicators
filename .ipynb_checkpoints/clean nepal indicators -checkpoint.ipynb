```python
# Add percentage sign to the specified column
nepal_data.loc[nepal_data['Indicator Name'].str.contains('%') & ~nepal_data['amount'].isnull(), 'amount'] = nepal_data['amount'].apply(lambda x: str(x) + '%' if pd.notna(x) else x)

# Convert the column back to numerical values, handling 'NaN' appropriately
nepal_data['amount'] = pd.to_numeric(nepal_data['amount'].str.rstrip('%'), errors='coerce')

# Display the updated DataFrame
nepal_data
```
